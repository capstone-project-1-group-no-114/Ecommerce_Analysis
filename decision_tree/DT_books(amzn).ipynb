```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df0 = pd.read_excel("z1 amazon_books_webscraping.xlsx")
df0.head()
```

```python
df0.isnull().sum()
df0.shape
```

```python
q = df0['DISCOUNT'].tolist()
q
dct = []
for i in q:
    if i[3] == "%":
        j = i[1:3]
        k = float(j)
        dct.append(k)
    elif i[2] == "%":
        d = i[1:2]
        s = float(d)
        dct.append(s)
dct
```

```python
w = df0['RATING'].tolist()
w
rt = []
for i in w:
    if i[17] == i[-1]:
        j = i[0:3]
        k = float(j)
        rt.append(k)
    elif i[16] == i[-1]:
        j = i[0]
        k = float(j)
        rt.append(k)
rt
```

```python
e = df0['NUMBER OF RATING'].tolist()
e
nor = []
for i in e:
    if "," in i:
        j = i.replace(",","")
        k = float(j)
        nor.append(k)
    else:
        j = float(i)
        nor.append(j)
nor
```

```python
avg_nor = sum(nor)/len(nor)
avg_nor
```

```python
demand = []
for i in range(len(rt)):
        if rt[i]>4.5 and nor[i]>avg_nor:
            demand.append("high demand")
        elif rt[i]>4.5 and nor[i]<avg_nor:
            demand.append("medium demand") 
        else:
            demand.append("low demand")
demand
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
clf = DecisionTreeClassifier()
```

```python
x = pd.DataFrame({"ratings":rt, "num_ratings":nor})
y = demand
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state = 42)
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
